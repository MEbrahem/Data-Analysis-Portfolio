## Introduction

### Dataset Description 

>A person makes a doctor appointment, receives all the instructions and no-show. Who to blame? 110.527 medical appointments its 14 associated variables (characteristics). The most important one if the patient show-up or no-show to the appointment.

### Columns Description

>**PatientId:** Identification of a patient.
<br>**AppointmentID:** Identification of each appointment.
<br>**Gender:** Male or Female.
<br>**ScheduledDay:** The day someone called or registered the appointment, this is before appointment of course.
<br>**AppointmentDay:** The day of the actuall appointment, when they have to visit the doctor.
<br>**Age:** How old is the patient.
<br>**Neighbourhood:** Where the appointment takes place.
<br>**Scholarship:** True of False. Scholarship here describes a social welfare program offered by the Government of Brazil to provide financial aid to poor Brazilian families on the condition that the families ensure their children attend schools and get vaccinated. '1' means such a patient has the scholarship while '0' means the patient doesnt have the scholarship.
<br>**Hipertension:** True of False.
<br>**Diabetes:** True of False.
<br>**Alcoholism:** True of False.
<br>**Handcap:** True of False. Is the patient handicapped? Handicap 0 means the patient doesn't have a handicap. 1 means they do have a handicap. 2 and 3 are errors (invalid values).
<br>**SMS_received:** 1 or more messages sent to the patient.
<br>**No-show:** True or False. it shows if the patients showed up for their appointment or not. 'No': the patient showed up for the appointment, 'Yes': the patient didnt show up for their appointment.

### Data Cleaning

>1- Change columns names to lowers letters.
<br> 2- There's one row has Age value -1 and this is not logical so we will going to drop it.
<br> 3- Change ScheduledDay and AppointmentDay types to Datetime.
<br> 4- Change No-show column name to show.
<br> 5- Reverse data in show column.
<br> 6- Drop values (2,3,4) from handicap column.
<br> 7- convert columns patientid, appointmentid to str formate.
<br> 8- Extract hour, day name, month name and year from ScheduledDay and AppointmentDay columns.
<br> 9- Add a new column (waiting_days) to calculate the diffrence between ScheduledDay and AppointmentDay.
<br> 10- Replce M and F in gender column with Male and Female.
<br> 11- Replace 0 and 1 with No and Yes in columns(scholarship, hipertension, diabetes, alcoholism, handicap, sms_received).

#### 1- Change columns names to lowers letters.

```python
data.columns=data.columns.str.lower()
```

```python
show_columns()
```

#### 2- There's one row has Age value -1 and this is not logical so we will going to drop it

```python
ind_row_not_logical=data[data["age"]<0].index
```

```python
data.drop(ind_row_not_logical,inplace=True)
```

```python
data[data['age']==0].shape
```

```python
data[data['age']==0].sample(10)
```

>As it turns out, there are 3539 with the age value of 0, and it is also noted that the values of scholarship, hipertension, diabetes, alcoholism, and handicap are also 0. So I think it is best to remove these rows.

```python
ind_row_Age_0=data[data['age']==0].index
```

```python
data.drop(ind_row_Age_0,inplace=True)
```

```python
data['age'].describe()
```

#### 3- Change ScheduledDay and AppointmentDay types to Datetime.

```python
data[['scheduledday','appointmentday']].dtypes
```

```python
data['scheduledday']=pd.to_datetime(data['scheduledday'])
```

```python
data['appointmentday']=pd.to_datetime(data['appointmentday'])
```

```python
print(data[['scheduledday','appointmentday']].head())
print(data[['scheduledday','appointmentday']].dtypes)
```

#### 4- Change No-show column name to show.

```python
data.rename(columns={"no-show":"show"},inplace=True)
```

#### 5- Reverse data in show column.

```python
data['show'].replace({"No":"Show","Yes":"No-Show"},inplace=True)
```

#### 6- Drop values (2,3,4) from handcap column.

```python
indxrows_handcaphave_234=data[(data['handcap']==2) | (data['handcap']==3) | (data['handcap']==4)].index
```

```python
data.drop(indxrows_handcaphave_234,inplace=True)
```

#### 7- convert columns patientid, appointmentid to str formate

```python
data[['patientid','appointmentid']]=data[['patientid','appointmentid']].astype('str')
```

```python
data[['patientid','appointmentid']].dtypes
```

#### 8- Extract hour, day name, month name and year from ScheduledDay and AppointmentDay columns.

```python
data['schedul_hour']=data['scheduledday'].dt.hour.astype(str)
```

```python
data['schedul_day']=data['scheduledday'].dt.day_name()
```

```python
data['schedul_month']=data['scheduledday'].dt.month_name()
```

```python
data['schedul_year']=data['scheduledday'].dt.year.astype(str)
```

```python
data['appoint_hour']=data['appointmentday'].dt.hour.astype(str)
```

```python
data['appoint_day']=data['appointmentday'].dt.day_name().astype(str)
```

```python
data['appoint_month']=data['appointmentday'].dt.month_name().astype(str)
```

```python
data['appoint_year']=data['appointmentday'].dt.year.astype(str)
```

```python
data.info()
```

#### 9- Add a new column (waiting_days) to calculate the diffrence between ScheduledDay and AppointmentDay.

```python
data['waiting_days']=(data['appointmentday']-data['scheduledday']).dt.days.astype(int)
```

#### 10- Replce M and F in gender column with Male and Female.

```python
data['gender'].replace({"M":"Male","F":'Female'},inplace=True)
```

#### 11- Replace 0 and 1 with No and Yes in columns(scholarship, hipertension, diabetes, alcoholism, handicap, sms_received)

```python
data[['scholarship', 'hipertension', 'diabetes', 'alcoholism', 'handcap', 'sms_received']]=data[['scholarship', 'hipertension', 'diabetes', 'alcoholism', 'handcap', 'sms_received']].replace({0:"No",1:'Yes'})
```

```python
data.sample(10)
```
