You will analyze a dataset of the global average sea level change since 1880. You will use the data to predict the sea level change through year 2050.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

Use the data to complete the following tasks:

- Use Pandas to import the data from epa-sea-level.csv.

```python
df=pd.read_csv('epa-sea-level.csv')
```

```python
df.head()
```

```python
df.columns
```

Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axix.

```python
plt.figure(figsize=(12,8))
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'])
plt.title('Annual increase rate',size=20)
```

Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python
import scipy as sp
```

```python
res=sp.stats.linregress(x=df['Year'],y=df['CSIRO Adjusted Sea Level'])
```

```python
res
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'],label='Original Data',c='green')
plt.plot(df['Year'],res.intercept+res.slope*df['Year'],label='fitted line',linewidth=3)
plt.title('Best fit line',size=15)
plt.legend(prop={'size':10})
```

Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2013.

```python
predict_period=range(2013,2051)
```

```python
plt.figure(figsize=(12,8))
plt.plot(predict_period,res.intercept+res.slope*predict_period,label='Predict line',c='red')
plt.title('Predict Line through 2013-2050',size='15')
```

Final plot, The x label should be Year, the y label should be Sea Level (inches), and the title should be Rise in Sea Level.

```python
plt.figure(figsize=(10,5))
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'],label='Original Data',c='green')
plt.plot(df['Year'],res.intercept+res.slope*df['Year'],label='fitted line',linewidth=3)
plt.plot(predict_period,res.intercept+res.slope*predict_period,label='Predict line',c='red')
plt.legend(prop={'size':10})
plt.title('Rise in Sea Level',size=20)
```

From visualization of the dataset, we were able to predict sea level through 2050
